```python
#grouping column into categories
ranges_age = [(0, 20), (21, 40), (41, 60), (61, 80), (81, 100)]
categories_age = [' Young', 'Adult', 'Mid-Age', 'Retired', 'Old']
dt['age_category'] = pd.cut(dt['Age'], bins=[range[0] - 0.5 for range in ranges_age] + [max(range[1] for range in ranges_age) + 0.5], labels=categories_age)
ranges_hours = [(0, 40), (40, 60), (60, 80), (80,100)]
categories_hours = ['Normal', 'Hardwork', 'Extra-HardWork', 'Over-the-limit']
dt['hours_category'] = pd.cut(dt['hours-per-week'], bins=[range[0] - 0.5 for range in ranges_hours] + [max(range[1] for range in ranges_hours) + 0.5], labels=categories_hours)
```

```python
features=['Age','workclass','fnlwgt','sex','capital-gain','capital-loss','hours-per-week']
target=['income']
```

```python
#Capital-gain categorization
bins = [0, 5000, 10000, 15000, 20000, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High', 'Extremely High']
dt['capital-gain_category'] = pd.cut(dt['capital-gain'], bins=bins, labels=labels, right=False)
#Capital-loss categorization
bins = [0, 1000, 2000, 3000, 4000, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High', 'Extremely High']
dt['capital-loss_category'] = pd.cut(dt['capital-loss'], bins=bins, labels=labels, right=False)
#finalwgt categorization
bins = [0, 50000, 100000, 150000, 200000, float('inf')]
labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']
dt['fnlwgt_category'] = pd.cut(dt['fnlwgt'], bins=bins, labels=labels, right=False)
```

```python
dt
```

```python
#encoding
from sklearn.preprocessing import LabelEncoder
categorical_columns = ['capital-gain_category','capital-loss_category','fnlwgt_category', 'education-num', 'sex', 'age_category','hours_category','income','workclass']
label_encoder = LabelEncoder()
for col in categorical_columns:
    dt[col + '_encoded'] = label_encoder.fit_transform(dt[col])
```

```python
dt
```

```python
#KNN model
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
X = dt[['capital-gain_category_encoded','capital-loss_category_encoded','fnlwgt_category_encoded', 'education-num_encoded', 'sex_encoded', 'age_category_encoded','hours_category_encoded','workclass_encoded']]
y = dt['income_encoded']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

knn = KNeighborsClassifier(n_neighbors=15)  

knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
y_pred=knn.predict(X_test)

cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#Heatmap
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#trying out alternative methods to check the effect on accuracy
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


knn_hamming = KNeighborsClassifier(n_neighbors=15, metric='hamming')
knn_manhattan = KNeighborsClassifier(n_neighbors=15, metric='manhattan')
knn_minkowski = KNeighborsClassifier(n_neighbors=15, metric='minkowski', p=2)  


knn_hamming.fit(X_train, y_train)
knn_manhattan.fit(X_train, y_train)
knn_minkowski.fit(X_train, y_train)

y_pred_hamming = knn_hamming.predict(X_test)
y_pred_manhattan = knn_manhattan.predict(X_test)
y_pred_minkowski = knn_minkowski.predict(X_test)

print("Classification Report - Hamming Distance:")
print(classification_report(y_test, y_pred_hamming))

print("Classification Report - Manhattan Distance:")
print(classification_report(y_test, y_pred_manhattan))

print("Classification Report - Minkowski Distance (Euclidean):")
print(classification_report(y_test, y_pred_minkowski))
```
